```python
x=df[['City_Tier','Occupation','Dependents','Income','Loan_Repayment','Insurance','Rent']]
y=df['Groceries']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb1=LabelEncoder()
lb2=LabelEncoder()
```

```python
df['City_Tier']=lb1.fit_transform(df['City_Tier'])
df['Occupation']=lb2.fit_transform(df['Occupation'])   
```

```python
import pickle
pickle.dump(lb1,open('Groceries_City.pkl','wb')) 
```

```python
pickle.dump(lb2,open('Groceries_Occupation.pkl','wb')) 
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr.score(x_train,y_train)
```

```python
lr.intercept_
```

```python
from sklearn.model_selection import cross_val_score

```
